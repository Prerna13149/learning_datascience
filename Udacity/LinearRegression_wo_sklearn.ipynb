```python
def linearReg(train_X, train_Y, alpha):
    #print(train_X.shape)
    #print(train_Y.shape)
    #n = len(train_X)
    m = 0.5
    b = 0.5
    for i in range(len(train_X)):
        #print(train_X.iloc[i]*m + b)
        print("b: ", b)
        print("m: ", m)
        y_pred = train_X.iloc[i]*m + b
        #print(y_pred[0])
        #print(train_Y.loc[i,1])
        #print(train_Y.iloc[i]-y_pred)
        m = m + (alpha * (train_Y.loc[i,1] - y_pred)*train_X.iloc[i])
        b = b + (alpha * (train_Y.loc[i,1] - y_pred))
        print("-----------")
    return m, b
```

```python
print(len(data))
data_x = data[[0]]#iloc[:,0]
data_y = data[[1]]#iloc[:,1]
print(len(data_x))
#print(data_y)
```

```python
m, b = linearReg(data_x, data_y, 0.5)
```
